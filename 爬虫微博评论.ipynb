```python
def parse_one_page(html):
    soup=BeautifulSoup(html,'lxml')
    items=soup.find_all(class_='ctt')
    for item in items:
        item = remove_tags(str(item)) #删除内部无用标签
        pattern = re.compile('.*?:(.*)',re.S) #删除回复@
        if re.findall(pattern,item):
            item = ''.join(re.findall(pattern,item)) #合并成字符串形式
        print(item)
        write_to_file(item)
```

### 结巴分词

```python
import jieba.analyse
import csv
import codecs

path = 'pinglun1.txt'
file_in = open(path, 'r',encoding="utf-8")
content = file_in.read()

try:
    jieba.analyse.set_stop_words('table.txt') #中文词库
    tags = jieba.analyse.extract_tags(content, topK=200, withWeight=True)
    for v, n in tags:
        #权重是小数，为了凑整，乘了一万
        file=open('a.txt', 'a', encoding='utf-8')
        file.write(v+'\t'+str(int(n * 10000)))
        file.write('\n')
    file.close()
finally:
    file_in.close()
```
