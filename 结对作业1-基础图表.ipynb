## 观看设备

```python
def c(data):
    y = [x for x in data['观看设备'].str.split('┋')]
    tel = 0
    comp = 0
    for item in y:
        for select in item:
            if select[0]=='手':
                tel = tel+1
            else:
                comp = comp + 1
    a=[]
    a.append(tel)
    a.append(comp)
    return a
```

```python
plt.figure(figsize=(4,4))
watch_labels= ['手机/平板','电脑']
count=69
share =[(c(data))[0]/count,(c(data))[1]/count]
explode = [0.04, 0.04]
plt.pie(share, explode = explode,
        labels = watch_labels, autopct = '%3.2f%%',
        startangle = 60, shadow = False,
        colors = ['lightgreen', 'lightskyblue'])
plt.title('观看设备倾向',size=18)
plt.axis('equal')
plt.savefig('观看设备.jpg', dpi=300) 
plt.show()
```

## 线上授课优点

```python
def c2(data):
    y = [x for x in data['线上授课优点'].str.split('┋')]
    a1 = 0
    a2 = 0
    a3 = 0
    a4 = 0
    a5 = 0
    for item in y:
        for select in item:
            if select[0]=='时':
                a1+=1
            elif select[0]=='不':
                a2+=1
            elif select[0]=='支':
                a3+=1
            elif select[0]=='师':
                a4+=1
            else:
                a5+=1
    a=[]
    a.extend([a1,a2,a3,a4,a5])
    return a
```

```python
plt.figure(figsize=(8,5)) 
labels8 = ['时间分配自由','不受老师管理，想干什么干什么','支持录屏，可以重复观看','师生互动不会显得环境吵杂','其他']
share =[(c2(data))[0]/count,(c2(data))[1]/count,(c2(data))[2]/count,(c2(data))[3]/count,(c2(data))[4]/count]
x=np.arange(len(labels8))
bar_width=0.5
plt.barh(x,share,bar_width)
plt.title('线上授课优点',size=18)
for a,b in zip(x,share): plt.text(b+0.03,a-0.05,"%.1f%%"%(b*100),ha='center', va= 'bottom',fontsize=12)
plt.yticks(x,labels8)
plt.savefig('线上授课优点.jpg', dpi=300) 
plt.show()
```

## 线上授课缺点

```python
def c3(data):
    y = [x for x in data['线上授课缺点'].str.split('┋')]
    a1 = 0
    a2 = 0
    a3 = 0
    a4 = 0
    a5 = 0
    for item in y:
        for select in item:
            if select[0]=='平':
                a1+=1
            elif select[0]=='注':
                a2+=1
            elif select[0]=='知':
                a3+=1
            elif select[0]=='周':
                a4+=1
            else:
                a5+=1
    a=[]
    a.extend([a1,a2,a3,a4,a5])
    return a
```

```python
plt.figure(figsize=(8,5)) 
labels9=['平台声音杂、卡顿','注意力难以集中','知识难以吸收','周边环境影响大','其他']
share =[(c3(data))[0]/count,(c3(data))[1]/count,(c3(data))[2]/count,(c3(data))[3]/count,(c3(data))[4]/count]
x=np.arange(len(labels9))
bar_width=0.5
plt.barh(x,share,bar_width)
plt.title('线上授课缺点',size=18)
for a,b in zip(x,share): plt.text(b+0.04,a-0.05,"%.1f%%"%(b*100),ha='center', va= 'bottom',fontsize=12)
plt.yticks(x,labels9)
plt.savefig('线上授课缺点.jpg', dpi=300) 
plt.show()
```
