```python
import pandas as pd
import numpy as np
from pandas import Series
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(10,5))
plt.rcParams['font.sans-serif'] = ['simhei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
data = pd.read_excel('data.xls')
```

## 联合分析

### 做笔记情况与听课专注力比较

```python
def c4(data):
    for i in range(len(data)):
        line=data.iloc[i,:]
        if line['线上听课做笔记情况']=='一直做笔记':
            line['线上听课做笔记情况']=3
        elif line['线上听课做笔记情况']=='适当做笔记':
            line['线上听课做笔记情况']=2
        else:
            line['线上听课做笔记情况']=1
```

```python
c4(data)
```

```python
data['线上听课做笔记情况']=[int(f) for f in data['线上听课做笔记情况']]
data['听课专注力']=data['听课专注力'].astype('category')
```

```python
plt.yticks(range(4),('','根本不做','适当做笔记','一直做笔记'))
sns.violinplot(x="听课专注力",y="线上听课做笔记情况",data=data)
plt.tight_layout() #图片保存完整
plt.savefig('1.jpg', dpi=300)
plt.show()
```

听课过程中，大多同学难以避免会走神，这部分偶尔走神的同学也会适当地做些笔记，但也有只是专注听课但不做笔记的人。
认真专注的同学都会选择做一些笔记，或多或少。
根本没听的同学自然就不存在做笔记这一选择咯。

### 课堂知识吸收度和听课专注力比较

```python
def c5(data):
    for i in range(len(data)):
        line=data.iloc[i,:]
        if line['课堂知识的吸收度']=='0-25%':
            line['课堂知识的吸收度']=1
        elif line['课堂知识的吸收度']=='26%-50%':
            line['课堂知识的吸收度']=2
        elif line['课堂知识的吸收度']=='51%-75%':
            line['课堂知识的吸收度']=3
        else:
            line['课堂知识的吸收度']=4
```

```python
c5(data)
```

```python
data['课堂知识的吸收度']=[int(f) for f in data['课堂知识的吸收度']]
```

```python
plt.yticks(range(5),('','0-25%','26%-50%','51%-75%','76%-100%'))
sns.violinplot(x="听课专注力",y="课堂知识的吸收度",data=data)
plt.tight_layout()
plt.savefig('2.jpg', dpi=300)
plt.show()
```

听课过程中，偶尔走神的同学对于课堂知识的吸收度大多在51%-75%，同时也有部分吸收度在26%-50%和76%-100%的。
根本没听的同学显然对于课堂知识的吸收度普遍偏低，大多都是0-25%。
认真专注的同学明显吸收能力在76%-100%。
显然上课时同学们的专注对于课堂知识的吸收度起着重要的作用。

### 课堂知识吸收度与作业完成度比较

```python
def c6(data):
    for i in range(len(data)):
        line=data.iloc[i,:]
        if line['作业完成度']=='顺便做做':
            line['作业完成度']=1
        elif line['作业完成度']=='能完成，但是百度成分高':
            line['作业完成度']=2
        elif line['作业完成度']=='能独立完成':
            line['作业完成度']=3
```

```python
c6(data)
```

```python
data['课堂知识的吸收度']=data['课堂知识的吸收度'].astype('category')
data['作业完成度']=[int(f) for f in data['作业完成度']]
```

```python
plt.yticks(range(4),('','顺便做做','能完成，但是水分大','能独立完成'))
sns.violinplot(x="课堂知识的吸收度",y="作业完成度",data=data)
plt.savefig('3.jpg', dpi=300)
plt.show()
```

对于课堂知识吸收度不理想的部分同学，对作业的态度还是很好的，能够完成老师布置的作用，只有少数人只是敷衍一下。
吸收度能达到26%-50%的同学都能很好的完成作业，但还是很多人并不能独立完成，需要借助于百度一类的外力。
吸收度中等偏上的同学很多都能独立完成作业，但也有人自持听懂了，就随便做下作业。
吸收度很好的同学几乎都能独立完成作业，也有部分同学借助了外力。
可见，对于课堂知识的吸收能力在一定程度上，与作业的完成度是有关系的，但主要还是态度问题。
并不是说难以理解课堂知识的人对于作业就会轻视，听懂了的同学也同样会敷衍作业。
