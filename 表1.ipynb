### 知识吸收度

```python
dataAbs=data['知识吸收度'].value_counts(ascending=False)
absolute=['很低','低','一般','高']
count=data['知识吸收度'].count()
share =[dataAbs['0%-25%']/count,dataAbs['25%-50%']/count,dataAbs['50%-75%']/count,dataAbs['75%-100%']/count]
explode = [0.04, 0.04,0.04,0.04]
plt.pie(share, explode = explode,
        labels = absolute, autopct = '%3.2f%%',
        startangle = 0, shadow = False,
        colors = ['lightgreen', 'lightskyblue','#32CD32','#38B0DE'])
plt.title('学生网课知识吸收程度',size=18)
plt.axis('equal')
plt.savefig('学生网课知识吸收程度.jpg', dpi=300) 
plt.show()
```

### 学习场所

```python
def b(data):
    y = [x for x in (data['学习场所'].dropna()).str.split('┋')]
    m = 0
    n = 0
    for item in y:
        for select in item:
            if select == '被窝':
                m += 1
            elif select == '桌前':
                n += 1
    z=[]
    z.extend([m,n])
    return z
```

```python
space=['被窝','桌前']
count=data['学习场所'].count()
share =[b(data)[0]/count,b(data)[1]/count]
explode = [0.04, 0.04]
plt.pie(share, explode = explode,
        labels = space, autopct = '%3.2f%%',
        startangle = 0, shadow = False,
        colors = ['lightgreen', 'lightskyblue'])
plt.title('学生网课学习场所',size=18)
plt.axis('equal')
plt.savefig('学生网课学习场所.jpg', dpi=300) 
plt.show()
```

### 观看设备统计

```python
def a(data):
    y = [x for x in (data['学习设备'].dropna()).str.split('┋')]
    tel = 0
    comp = 0
    for item in y:
        for select in item:
            if select == '手机/平板':
                tel += 1
            elif select == '电脑':
                comp += 1
    z=[]
    z.extend([tel,comp])
    return z
```

```python
plt.figure(figsize=(4,4))
watch_labels= ['手机/平板','电脑']
count=data['学习设备'].count()
share =[(a(data))[0]/count,(a(data))[1]/count]
explode = [0.04, 0.04]
plt.pie(share, explode = explode,
        labels = watch_labels, autopct = '%3.2f%%',
        startangle = 90, shadow = False,
        colors = ['lightgreen', 'lightskyblue'])
plt.title('学生网课观看设备',size=18)
plt.axis('equal')
plt.savefig('学生网课观看设备.jpg', dpi=300) 
plt.show()
```
